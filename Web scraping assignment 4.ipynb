# 5. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used

```python
driver=webdriver.Chrome(r"C:\Users\jeeves training\chromedriver.exe")
driver.get('https://github.com/')
WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary'))).click()
WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a'))).click()
#driver.maximize_window()
```

```python
#Obtaining urls
urls=[]
for i in driver.find_elements_by_xpath('//*[@id="js-pjax-container"]/div[3]/div/div[2]/article/h1/a'):
    urls.append(i.get_attribute("href"))
```

```python
Title=[]
Description=[]
contributer_count=[]
Language_used=[]
#scrapping data
for i in urls:
    driver.get(i)
    
    #Title
    try:
        title=driver.find_element_by_xpath('//h1[@class=" d-flex flex-wrap flex-items-center break-word f3 text-normal"]')
        Title.append(title.text.replace('\n',' '))
    except NoSuchElementException:
        Title.append('-')
        
    #Description
    try:
        description=driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div[1]/div/p')
        Description.append(description.text)
    except NoSuchElementException:
        Description.append('-')
        
    #Contributors count
    try:
        count=driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div[5]/div/h2/a/span')
        contributer_count.append(count.text)
    except NoSuchElementException:
        contributer_count.append('-')
    #Language used
    try:
        lang=driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div[6]/div/ul/li/a/span[1]')
        Language_used.append(lang.text.replace('\n',' '))
    except NoSuchElementException:
        Language_used.append('-')
        
print(len(Title),len(Description),len(contributer_count),len(Language_used))
```

```python
Github_repository=pd.DataFrame({})
Github_repository['Repository_Title']=Title
Github_repository['Repository_Description']=Description
Github_repository['Contributer_Count']=contributer_count
Github_repository['Language_Used']=Language_used
Github_repository
```

```python
driver=webdriver.Chrome(r"C:\Users\jeeves training\chromedriver.exe")
driver.get('https://archive.ics.uci.edu/ml/index.php')
driver.maximize_window()
#loc = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td/span/b/a").click()
WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '/html/body/table[2]/tbody/tr/td/span/b/a'))).click()

Dataset=[]
dataset = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a")
for i in dataset:
    if i.text is None:
        Dataset.append("--")
    else:
        Dataset.append(i.text)

Datatype=[]
datatype = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p")
for i in datatype:
    if i.text != 'Data Types':
        Datatype.append(i.text)
```

```python
Default_task=[]
task = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]/p")
for i in task:
    if i.text != 'Default Task':
        Default_task.append(i.text)
```

```python
Attribute_type=[]
attribute = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p")
for i in attribute:
    if i.text != 'Attribute Types':
        Attribute_type.append(i.text)
```

```python
Number_of_instance=[]
instance = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]/p")
for i in instance:
    if i.text != '# Instances':
        Number_of_instance.append(i.text)
```

```python
Number_of_attributes=[]
att = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]/p")
for i in att:
    if i.text != '# Attributes':
        Number_of_attributes.append(i.text)
```

```python
Year=[]
year = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p")
for i in year:
    if i.text != 'Year':
        Year.append(i.text)
```

```python
DATA_SET=pd.DataFrame({})
DATA_SET['DATASET']=Dataset
DATA_SET['DATATYPE']=Datatype
DATA_SET['DEFAULT_TASK']=Default_task
DATA_SET['ATTRIBUTE_TYPE']=Attribute_type
DATA_SET['NUMBER_OF_INSTANCE']=Number_of_instance
DATA_SET['NUMBER_OF_ATTRIBUTES']=Number_of_attributes
DATA_SET['YEAR']=Year
DATA_SET
```
