### Q2:In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a dataframe and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”.     In case, if any of the details are missing for any of the product then replace it by “-“.

```python
brand=[]
name=[]
rating=[]
num_ratings=[]
price=[]
return_exchange=[]
delivery=[]
availability=[]
other_details=[]
prod_urls=[]
for page in range(0,2):
    urls=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
    for i in urls:
        prod_urls.append(i.get_attribute("href"))
    button=driver.find_element_by_xpath("//li[@class='a-last']/a")
```

```python
for i in prod_urls:
    driver.get(i)
    try:#Brand
        item_brand=driver.find_element_by_xpath("//[@id='productTitle']")
        brand.append(item_brand.text)
    except NoSuchElementException:
        brand.append('--')
    try:#Name
        item_name=driver.find_element_by_xpath("//td[@class='a-size-base prodDetAttrValue'][12]")
        name.append(item_name.text)
    except NoSuchElementException:
        name.append('--')
    try:#Price
        item_price=driver.find_element_by_id("priceblock_ourprice")
        price.append(item_price.text)
    except NoSuchElementException:
        price.append('--')
    try:#Rating
        item_rating=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        rating.append(item_rating.text)
    except NoSuchElementException:
        rating.append('--')
    try:#Number Rating
        item_num_rating=driver.find_element_by_xpath("//span[@class='a-size-base']")
        num_ratings.append(item_num_rating.text)
    except NoSuchElementException:
        num_ratings.append('--')
    try:#return exchange
        item_return=driver.find_element_by_xpath("//a[@class='a-size-small a-link-normal a-text-normal']")
        return_exchange.append(item_return.text)
    except NoSuchElementException:
        return_exchange.append('--')
    try:#Delivery
        item_delivery=driver.find_element_by_xpath("//div[@class='a-section a-spacing-none']/b")
        delivery.append(item_delivery.text)
    except NoSuchElementException:
        delivery.append('--')
    try:#Availability
        item_a=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-sucess']")
        availability.append(item_a.text)
    except NoSuchElementException:
        availability.append('--')
    try:#other details
        item_ot=driver.find_element_by_xpath("//div[@class='a-section a-spacing-medium a-spacing-top-small']")
        other_details.append(item_ot.text)
    except NoSuchElementException:
        other_details.append('--')
```

```python
product_details=pd.DataFrame({})
#product_details['Brand']=brand
product_details['Name']=name
product_details['Price']=price
product_details['Rating']=rating
product_details['Number_Ratings']=num_ratings
product_details['Return_Exchange']=return_exchange
product_details['Delivery']=delivery
product_details['Availability']=availability
product_details['Other_Details']=other_details
product_details['product_url']=prod_urls
product_details
```

# q3:Write a python program to access the search bar and search button on images.google.com and scrape 100 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’.

```python
driver=webdriver.Chrome(r"C:\Users\jeeves training\chromedriver.exe")
driver.get('https://images.google.com/?gws_rd=ssl')

search_product=driver.find_element_by_xpath('//input[@class="gLFyf gsfi"]')

search_product.clear()
search_string = input("Input the string you want to search for:")
search_product.send_keys(search_string)
search_loc = driver.find_element_by_xpath('//div[@class="FAuhyb"]').click()
```

```python
imgResults=[]
imgResults = driver.find_elements_by_xpath("//img[contains(@class,'rg_i Q4LuWd')]")
totalResults=len(imgResults)
```

```python
img_urls = set()
for i in  range(0,len(imgResults)):
    img=imgResults[i]
    try:
        img.click()
        time.sleep(2)
        actual_images = driver.find_elements_by_css_selector('img.n3VNCb')
        for actual_image in actual_images:
            if actual_image.get_attribute('src') and 'https' in actual_image.get_attribute('src'):
                img_urls.add(actual_image.get_attribute('src'))
    except ElementClickInterceptedException or ElementNotInteractableException as err:
        print(err)
```

```python
import os
import requests
import io
from PIL import Image
os.chdir('C:\RAJESH\DATA TRAINED\PROJECTS')
baseDir=os.getcwd()
```

```python
for i, url in enumerate(img_urls):
    file_name = f"{i:150}.jpg"    
    try:
        image_content = requests.get(url).content

    except Exception as e:
        print(f"ERROR - COULD NOT DOWNLOAD {url} - {e}")

    try:
        image_file = io.BytesIO(image_content)
        image = Image.open(image_file).convert('RGB')
        
        file_path = os.path.join(baseDir, file_name)
        
        with open(file_path, 'wb') as f:
            image.save(f, "JPEG", quality=85)
        print(f"SAVED - {url} - AT: {file_path}")
    except Exception as e:
        print(f"ERROR - COULD NOT SAVE {url} - {e}")
```

### 4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
driver=webdriver.Chrome(r"C:\Users\jeeves training\chromedriver.exe")
```

```python
driver.get('https://www.flipkart.com/')
```

```python
search_tag = driver.find_element_by_xpath("//div[@class='_1cmsER']/form/div/div/input")
search_tag.clear()
search_tag.send_keys("pixel 4A")
driver.find_element_by_xpath("//button[@class='L0Z3Pu']").click()
```

```python
df=pd.DataFrame({})

brand_tags=driver.find_elements_by_xpath("//div[@class='_4rR01T']")
brand=[]
for i in brand_tags:
    brand.append(i.text)
df['mixedvalue']=brand
df['dummy1'] = df['mixedvalue'].str.split("(").str[0]
df['dummy2'] = df['mixedvalue'].str.split("(").str[1]

df['BranName'] = df['dummy1'].str.split().str[0]
df['PhoneName'] = df['dummy1'].str.split().str[1:3]
df['color'] = df['dummy2'].str.split(',').str[0]
```

```python
ram_rom=driver.find_elements_by_xpath("//ul[@class='_1xgFaf']//li[1]")
ram_rom_clubbed=[]
for i in ram_rom[:40]:
    ram_rom_clubbed.append(i.text)
df['ram_rom_clubbed']=ram_rom_clubbed
df['RAM'] = df['ram_rom_clubbed'].str.split("|").str[0]
df['ROM'] = df['ram_rom_clubbed'].str.split("|").str[1:3]
```

```python
display=driver.find_elements_by_xpath("//li[2][@class='rgWa7D']")
display_clubbed=[]
for i in display[:40]:
    display_clubbed.append(i.text)
df['display_clubbed']=display_clubbed
df['DisplaySize'] = df['display_clubbed'].str.split().str[0:4]
df['DisplayResolution'] = df['display_clubbed'].str.split(")").str[1].fillna('-')
```

```python
camera=driver.find_elements_by_xpath("//ul[@class='_1xgFaf']//li[3]")
camera_clubbed=[]
for i in camera[:40]:
    camera_clubbed.append(i.text)
df['camera_clubbed']=camera_clubbed
df['RearCamera'] = df['camera_clubbed'].str.split("|").str[0]
df['FrontCamera'] = df['camera_clubbed'].str.split("|").str[1].fillna('-')
```

```python
battery=driver.find_elements_by_xpath("//ul[@class='_1xgFaf']//li[4]")
battery_clubbed=[]
for i in battery[:40]:
    battery_clubbed.append(i.text)
df['battery']=battery_clubbed
```

```python
processor=driver.find_elements_by_xpath("//ul[@class='_1xgFaf']//li[5]")
processor_clubbed=[]
for i in processor:
    if i.text is None:
        processor_clubbed.append("-")
    else:
        processor_clubbed.append(i.text)
df['processor_clubbed']=processor_clubbed
df['Processor'] = df['processor_clubbed'].str.split().str[0:2]
df['ProcessorCore'] = df['processor_clubbed'].str.split().str[-2:].fillna('-')
```

```python
productURL=driver.find_elements_by_xpath("//a[@class='_1fQZEK']")
productURL_clubbed=[]
for i in productURL:
    if i.text is None:
        productURL_clubbed.append("-")
    else:
        productURL_clubbed.append(i.get_attribute('href'))
df['productURL']=productURL_clubbed
```

```python
delcolumns=['mixedvalue','dummy1','dummy2','ram_rom_clubbed','display_clubbed','camera_clubbed','processor_clubbed']
df.drop(delcolumns,axis=1,inplace=True)
```

```python
df.to_csv('Flipcart_Mobile_data.csv')
df.head()
```
